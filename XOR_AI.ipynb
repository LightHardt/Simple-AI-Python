# Import tensorflow as we will use keras for making model

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Only shows errors hides warning good for hiding retracing error shown when doing
# model.predict() later on
tf.get_logger().setLevel('ERROR')
```

# Create Model

```python
model = keras.Sequential()
model.add(keras.layers.Dense(2, activation='relu', input_dim=2))
model.add(keras.layers.Dense(8, activation='relu'))
model.add(keras.layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

# Create XOR Training and Validation Set

```python
training_x = [[1,1],[1,0],[0,1],[0,0]]
training_y = [[0],[1],[1],[0]]

X = np.array(training_x)
Y = np.array(training_y)
```

# Train model

```python
history = model.fit(X,Y,epochs=1000,batch_size=1)
model.save("XOR_Weights")
```

# Test Model

```python
print(model.predict(np.asarray([[0,0]])))
print(model.predict(np.asarray([[1,0]])))
print(model.predict(np.asarray([[0,1]])))
print(model.predict(np.asarray([[1,1]])))
```

# Here are different ways of vizualizing your model

```python
model.summary()
```

```python
tf.keras.utils.plot_model(model,show_shapes=True, show_layer_names=True, show_layer_activations=True)
```

```python
for layer in model.layers:
    print('Layer Name: ' + layer.name)
    print('Layer Type: ' + layer.__class__.__name__)
    print('Layer input shape: {}'.format(layer.input_shape))
    print('Layer output shape: {}'.format(layer.output_shape))
    print('Layer parameter count: {}'.format(layer.count_params()))
    print('Layer weights: \n{}'.format(layer.weights))
    print()
```

```python
from keras_visualizer import visualizer

visualizer(model, format='png',view=True)
```

# Visualize loss (learning)

```python
loss_values = history.history['loss']

epochs = range(1, len(loss_values)+1)

plt.plot(epochs, loss_values, label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

# Visualize accuracy

```python
accuracy = history.history['accuracy']

plt.plot(epochs, accuracy, label='Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```
