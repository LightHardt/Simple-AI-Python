# Imports

```python
import tensorflow
from tensorflow import keras
import numpy
import matplotlib.pyplot as plt
```

# Build model

```python
model = keras.Sequential()
model.add(keras.layers.Dense(2,activation='relu',input_dim=2))
model.add(keras.layers.Dense(4,activation='relu'))
model.add(keras.layers.Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

# Training Data

```python
training_data = [[0,0],[0,1],[1,0],[1,1]]
training_labels = [[1],[0],[0],[1]]

X = numpy.asarray(training_data)
Y = numpy.asarray(training_labels)
```

# Train model

```python
history = model.fit(X,Y,batch_size=1,epochs=1000)
model.save('XAND_weights')
```

# Test Model

```python
print(model.predict(numpy.asarray([[0,0]])))
print(model.predict(numpy.asarray([[0,1]])))
print(model.predict(numpy.asarray([[1,0]])))
print(model.predict(numpy.asarray([[1,1]])))
```

# Visualize Loss

```python
loss_values = history.history['loss']

epochs = range(1, len(loss_values)+1)

plt.plot(epochs, loss_values, label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

# Visualize Accuarcy

```python
accuracy = history.history['accuracy']

plt.plot(epochs, accuracy, label='Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```
